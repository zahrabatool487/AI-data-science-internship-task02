```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Load the dataset
print("=" * 60)
print("LOADING IRIS DATASET")
print("=" * 60)

# Load from seaborn's built-in dataset
iris = sns.load_dataset('iris')
print("Dataset loaded successfully!")

# Display basic dataset structure
print("\n" + "=" * 60)
print("DATASET STRUCTURE")
print("=" * 60)

# Display shape (rows, columns)
print(f"Dataset Shape: {iris.shape}")
print(f"Number of rows: {iris.shape[0]}")
print(f"Number of columns: {iris.shape[1]}")

# Display column names and data types
print("\nColumn Information:")
print(iris.dtypes)

# Display first few rows
print("\nFirst 5 rows of the dataset:")
print(iris.head())

# Display summary statistics
print("\n" + "=" * 60)
print("SUMMARY STATISTICS")
print("=" * 60)
print(iris.describe())

# Check for missing values
print("\nMissing Values Check:")
print(iris.isnull().sum())

# Species distribution
print("\nSpecies Distribution:")
print(iris['species'].value_counts())
```

```python
# ============================================================
# VISUALIZATION 1: Pairplot (Scatter Matrix)
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION 1: PAIRPLOT (RELATIONSHIPS BETWEEN VARIABLES)")
print("=" * 60)

# Create a pairplot to visualize relationships between all variables
# Scatter plots show relationships, histograms show distributions
pairplot = sns.pairplot(iris, hue='species', palette='viridis',
                        diag_kind='hist', markers=['o', 's', 'D'])
pairplot.fig.suptitle('Iris Dataset: Pairwise Relationships by Species',
                      y=1.02, fontsize=16)
plt.show()

# ============================================================
# VISUALIZATION 2: Scatter Plot (Specific Relationship)
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION 2: SCATTER PLOT (PETAL LENGTH VS PETAL WIDTH)")
print("=" * 60)

fig, ax = plt.subplots(figsize=(10, 6))
scatter = sns.scatterplot(data=iris, x='petal_length', y='petal_width',
                          hue='species', style='species', s=100,
                          palette='Set2', ax=ax)
ax.set_title('Iris Flowers: Petal Length vs Petal Width', fontsize=16)
ax.set_xlabel('Petal Length (cm)', fontsize=12)
ax.set_ylabel('Petal Width (cm)', fontsize=12)
ax.legend(title='Species', title_fontsize=12)
plt.show()

# ============================================================
# VISUALIZATION 3: Histograms (Data Distribution)
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION 3: HISTOGRAMS (FEATURE DISTRIBUTIONS)")
print("=" * 60)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Iris Flower Features', fontsize=16, y=1.02)

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

for idx, (feature, color) in enumerate(zip(features, colors)):
    ax = axes[idx // 2, idx % 2]
    sns.histplot(data=iris, x=feature, kde=True, color=color, bins=15, ax=ax)
    ax.set_title(f'Distribution of {feature.replace("_", " ").title()}', fontsize=14)
    ax.set_xlabel(f'{feature.replace("_", " ").title()} (cm)', fontsize=11)
    ax.set_ylabel('Frequency', fontsize=11)

plt.tight_layout()
plt.show()

# ============================================================
# VISUALIZATION 4: Box Plots (Outliers & Spread)
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION 4: BOX PLOTS (SPREAD & OUTLIERS)")
print("=" * 60)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Box Plots of Iris Features by Species', fontsize=16, y=1.02)

for idx, feature in enumerate(features):
    ax = axes[idx // 2, idx % 2]
    sns.boxplot(data=iris, x='species', y=feature, palette='Pastel1', ax=ax)
    ax.set_title(f'{feature.replace("_", " ").title()} by Species', fontsize=14)
    ax.set_xlabel('Species', fontsize=11)
    ax.set_ylabel(f'{feature.replace("_", " ").title()} (cm)', fontsize=11)

    # Add swarm plot for better data point visualization
    sns.swarmplot(data=iris, x='species', y=feature, color='black',
                  alpha=0.5, size=4, ax=ax)

plt.tight_layout()
plt.show()

# ============================================================
# VISUALIZATION 5: Correlation Heatmap
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION 5: CORRELATION HEATMAP")
print("=" * 60)

# Calculate correlation matrix
numeric_iris = iris.select_dtypes(include=[np.number])
correlation_matrix = numeric_iris.corr()

fig, ax = plt.subplots(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
                      center=0, linewidths=1, linecolor='black', ax=ax,
                      cbar_kws={'label': 'Correlation Coefficient'})
ax.set_title('Correlation Matrix of Iris Features', fontsize=16)
plt.show()
```
